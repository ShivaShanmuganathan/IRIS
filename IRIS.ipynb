# Importing the dataset

```r
data_frame = read.csv("Iris.csv")
```

# Output Variable Species has been modified to Numeric Data

```r
head(data_frame)
```

```r
matrix_corr <- cor(data_frame, method="pearson")
matrix_corr
```

# The Highly Correlated Input Variable to the Output Variable is Petal_Width

```r
install.packages("corrplot")
```

```r
library(corrplot)
corrplot(matrix_corr, method = "ellipse")
```

# We can observe that Petal_Width can be used to classify the Species of the Flower

```r
plot(data_frame$PetalWidthCm, data_frame$Species, xlab="Petal Width", ylab="Species", main="Species plotted against Petal Width")
abline(-0.23072, 1.02674) #Coefficients of intercept and PetalWidthCm
```


## Creating a Linear Model with Highly Correlated Variable and Output Variable

```r
model <- lm(data_frame$Species ~ data_frame$PetalWidthCm)
modelSummary <- summary(model)
```

```r
plot(modelSummary$residuals, ylab="Residuals")
abline(0,0)
```
